```python
json_list = list()
for f in os.listdir(file_dir):
    cur_file = os.path.join(file_dir,f)
    with open(cur_file) as json_file:
        json_data = json.load(json_file)
        
    json_list.append(json_data)
```

```python
df = pd.DataFrame(json_list)
```

```python
def clean_rd(release_date):
    samp = df.iloc[0].release_date[0][1:].lower().replace(',','')
    date_object = datetime.strptime(samp, '%b %d %Y')
    time_stamp = time.mktime(date_object.timetuple())
    return time_stamp
```

```python
df['time_stamp'] = df['release_date'].apply(clean_rd)
```
