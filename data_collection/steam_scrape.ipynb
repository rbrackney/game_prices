```python
import steamscrape as sts
import numpy as np
import json
import os
import time
```

```python
with open("johcar_game_ids.json") as json_file:
    json_data = json.load(json_file)
```

```python
appids = []
#json_data[u'games']
          #for k,v in json_data.iteritems():
for item in json_data[u'response'][u'games']:
    appids.append(item[u'appid'])
appids = sorted(appids, reverse=True)
```

```python
for a_i in appids:
    fname = 'steam_%d.json' % (a_i)
    fpath = 'steam_jsons/%s' % fname
    if not fname in os.listdir('steam_jsons/'):
        try:
            out_dict, out_hs, error_result = sts.extract_info_games(a_i)
            with open(fpath, 'w') as outfile:
                json.dump(out_dict, outfile)
        except:
            print('problem with %d' % a_i)
        time.sleep(2)
        
```
